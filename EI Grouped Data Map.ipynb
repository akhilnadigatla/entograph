```python
import numpy as np
import pandas as pd

df = pd.read_csv('table-sridhar.csv')
```

```python
import geopandas as gpd
from shapely.geometry import Point, Polygon
import matplotlib.pyplot as plt
```

```python
world_map = gpd.read_file('World_Countries__Generalized_.shp')
```

```python
crs = {'init': 'epsg:4326'}
geometry = [Point(xy) for xy in zip(df['Longitude'], df['Latitude'])]
geo_df = gpd.GeoDataFrame(df, crs = crs, geometry = geometry)
```

```python
fig, ax = plt.subplots(figsize = (20, 20))
world_map.plot(ax = ax, alpha = 0.4, color = 'grey')
geo_df.plot(column = 'EI',ax = ax, alpha = 0.5, legend = True, markersize = 0.1)
plt.title('EI Data Map', fontsize = 15, fontweight = 'bold')
plt.xlim(-150, 150)
plt.ylim(-100, 100)
plt.savefig('ei_data.png', dpi=1000)
plt.show()
```
